## Задание 1

```python
df = pd.read_csv('dataset_0.csv')
```

```python
df.info()
df.head(10)
```

```python
# рассчитаем К корреляции "независимых" переменных и цели 'class'
df.corr()['class'].abs().sort_values()
```

```python
df[['feature_4', 'feature_20', 'feature_13', 'feature_17', 'feature_3', 'class']]\
        .plot(by='class', kind='box', figsize=(20, 5))
```

## Задание 2

```python
df = pd.read_csv('dataset_1.csv')
```

```python
df.info()
df.head(10)
```

```python
# построим диаграмму рассеяния
sns.scatterplot(df, x='feature_1', y='feature_2', hue='target')
```

```python
# значения target 0/1 распределены по отдельным "окружностям"
# найдем порог разделения "кластеров" в зависимости от радиуса "кругов" (sqrt(x ** 2 + y ** 2)) 
```

```python
plt.scatter(range(len(df)), (df['feature_1'] ** 2 + df['feature_2'] ** 2) ** (1 / 2))
```

```python
# => за пороговое можно принять значение 0.9. проведем проверочный расчет
```

```python
df['target_new'] = (df['feature_1'] ** 2 + df['feature_2'] ** 2) ** (1 / 2) < 0.9
df['target_new'] = df['target_new'].astype(int)
(df['target_new'] == df['target']).all()
```
